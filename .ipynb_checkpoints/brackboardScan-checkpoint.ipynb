```python
import cv2
import numpy as np
from IPython import display
from matplotlib import pyplot as plt


def imshow(img, format=".jpg", **kwargs):
    """ndarray 配列をインラインで Notebook 上に表示する。
    """
    img = cv2.imencode(format, img)[1]
    img = display.Image(img, **kwargs)
    display.display(img)


img = cv2.imread("img/IMG_0325.JPG")
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower = (75, 0, 0)
upper = (100,128, 150)
bin_img = cv2.inRange(hsv, lower, upper)
## フィルタの設定
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (10, 10))
closing = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel)
imshow(closing)

#result = cv2.bitwise_and(img, img, mask=closing) # 元画像とマスクを合成
#imshow(result)
```

```python
contours, hierarchy = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
con = list(filter(lambda x: cv2.contourArea(x) > 2000, contours))
output = cv2.drawContours(img, con, -1, (0,255,0), 3)
imshow(output)
```
